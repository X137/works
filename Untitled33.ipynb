```python
import sys
from PyQt5 import QtCore, QtWidgets ,QtGui
from PIL import ImageGrab, Image
import numpy as np
import cv2
import pytesseract

class Ui_FirstWindow(object):
    def setupUi(self, FirstWindow):
        
        FirstWindow.setObjectName("Translation")
        FirstWindow.resize(400, 300)
        self.centralWidget = QtWidgets.QWidget(FirstWindow)
        self.centralWidget.setObjectName("centralWidget")
        self.pushButton = QtWidgets.QPushButton(self.centralWidget)
        self.pushButton.setGeometry(QtCore.QRect(110, 130, 191, 23))
        self.pushButton.setObjectName("pushButton")
        self.textbox = QtWidgets.QLineEdit(self.centralWidget)
        self.textbox.resize(280,40)
        FirstWindow.setCentralWidget(self.centralWidget)

        self.retranslateUi(FirstWindow)
        QtCore.QMetaObject.connectSlotsByName(FirstWindow)
        
    def retranslateUi(self, MainWindow):
        _translate = QtCore.QCoreApplication.translate
        MainWindow.setWindowTitle(_translate("FirstWindow", "FirstWindow"))
        self.pushButton.setText(_translate("FirstWindow", "Capture Image"))

    def LoadSecondWindow(self):
        SecondWindow = QtWidgets.QMainWindow()
        ui = Ui_SecondWindow()
        ui.setupUi(SecondWindow)
        SecondWindow.show()
        
class Ui_SecondWindow(QtWidgets.QMainWindow):
    
    def paintEvent(self, event):
        qp = QtGui.QPainter(self)
        qp.setPen(QtGui.QPen(QtGui.QColor('black'), 1))
        qp.setBrush(QtGui.QColor(128, 128, 255, 128))
        qp.drawRect(QtCore.QRect(self.begin, self.end))

    def mousePressEvent(self, event):
        self.begin = event.pos()
        self.end = self.begin
        self.update()

    def mouseMoveEvent(self, event):
        self.end = event.pos()
        self.update()

    def mouseReleaseEvent(self, event):
        
        print("tessttt1")
        self.close()
        
        x1 = min(self.begin.x(), self.end.x())
        y1 = min(self.begin.y(), self.end.y())
        x2 = max(self.begin.x(), self.end.x())
        y2 = max(self.begin.y(), self.end.y())
        #   作成する
        print("tessttt2")
        
        img = ImageGrab.grab(bbox=(x1, y1, x2, y2))
        img.save('capture.png')
        img = cv2.cvtColor(np.array(img), cv2.COLOR_BGR2RGB)
        filename = 'capture.png'
        path  = r'Tesseract-OCR\tesseract.exe'
        pytesseract.pytesseract.tesseract_cmd = path
        text = pytesseract.image_to_string(Image.open(filename), lang="jpn")
        print(text)
        
        
class Controller:

    def __init__(self):
        pass

    def Show_FirstWindow(self):

        self.FirstWindow = QtWidgets.QMainWindow()
        self.ui = Ui_FirstWindow()
        self.ui.setupUi(self.FirstWindow)
        self.ui.pushButton.clicked.connect(self.Show_SecondWindow)
        self.FirstWindow.show()
        
    def Show_SecondWindow(self):
        
        self.SecondWindow = Ui_SecondWindow()
        self.SecondWindow.resize(1600, 900)
        self.SecondWindow.setWindowTitle(' ')
        self.SecondWindow.begin = QtCore.QPoint()
        self.SecondWindow.end = QtCore.QPoint()
        self.SecondWindow.setWindowOpacity(0.3)
        QtWidgets.QApplication.setOverrideCursor(
             QtGui.QCursor(QtCore.Qt.CrossCursor)
        )
        self.SecondWindow.setWindowFlags(QtCore.Qt.FramelessWindowHint)
        
        self.ui = Ui_SecondWindow()
        self.SecondWindow.show()
        
    def Print(self):
        print('After 99 hours of trying out everything')

if __name__ == '__main__':
    app = QtWidgets.QApplication(sys.argv)
    Controller = Controller()
    Controller.Show_FirstWindow()
    sys.exit(app.exec_())
```
